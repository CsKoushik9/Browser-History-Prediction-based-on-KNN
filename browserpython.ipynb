```python
#Extracting the history from the browser
import browserhistory as bh
from collections import Counter
dictval=bh.get_browserhistory()
data=[]
for i in dictval:
    for j in dictval[i]:
        data.append(list(j))
for i in data:
    i[-1]=i[-1].split(' ')[0]
dict_data={i[-1]:[] for i in data} 
for i in data:
    dict_data[i[-1]].append(i[:-2]) 
#Dictionary consists of date wise data i.e consists of dates as keys
```

```python
#Data Preprocessing
#Extracting the websites alone and replacing them in the actual url area
import re
for i in dict_data:
    for j in dict_data[i]:
        if re.search('search/',j[0]):
            j[0]=re.split('search/',j[0])[0]+'search/'
        elif re.search('url/',j[0]):
            j[0]=re.split('url/',j[0])[0]+'url/'
        elif re.search('com/',j[0]):
            j[0]=re.split('com/',j[0])[0]+'com/'
        elif re.search('edu/',j[0]):
            j[0]=re.split('edu/',j[0])[0]+'edu/'
        elif re.search('org/',j[0]):
            j[0]=re.split('org/',j[0])[0]+'org/'
        elif re.search('in/',j[0]):
            j[0]=re.split('in/',j[0])[0]+'in/'
        elif re.search('accounts/',j[0]):
            j[0]=re.split('accounts/',j[0])[0]+'accounts/'
        elif re.search('net/',j[0]):
            j[0]=re.split('net/',j[0])[0]+'net/'

for i in dict_data:
    l=[j for j in dict_data[i]]
    dict_data[i]=[j[0] for j in l]
```

```python
#Replacing the website names with website and number of times it is visited date wise
for i in dict_data:
    l=[]
    for j in Counter(dict_data[i]).most_common():
        l.append([j[0],j[1]])
    dict_data[i]=[j for j in l]
```

```python
#converting the date based data to website based data
l=[]
for i in dict_data:
    for j in dict_data[i]:
        l.append(j[0])
l=list(set(l))
dict_data1={i:[] for i in l}
```

```python
#dict_data1 contains the trained data of the model
for i in dict_data:
    for j in dict_data[i]:
        dict_data1[j[0]].append([i,j[1]])
print(dict_data1)
```

```python
#Prediction is done based on KNN but not actual KNN 
'''The difference between the dates and the number of times the website is visited on that date 
and the trend is taken into account'''
from datetime import datetime
import math
date_format = "%Y-%m-%d"
s=input('Enter the date in (YYYY-MM-DD) format: ')
s1=s.split('-')
if len(s1)==3:
    if len(s1[0])!=4 and len(s1[1])!=2 and len(s1[2])!=2:
        print('Invalid Format')
    else:
        dat=[]
        for i in dict_data1:
            m=[]
            su=0
            for j in dict_data1[i]:
                m.append((datetime.strptime(s,date_format)-datetime.strptime(j[0],date_format)).days)
                su+=1
            if min(m)<15:
                dat.append([i,math.floor((su+dict_data1[i][m.index(min(m))][1])/(len(j)+1))])
        print('Predicted Browser History dated ',s)
        print()
        for i in dat:
            if i[1]!=0:
                print('\tWebsite:',i[0])
                print('Number of times:',i[1])
                print()
else:
    print('Invalid Format')
```
